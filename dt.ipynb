<H3> DECISION TREE </H3>

```python
import pandas as pd
# extracting data into a df
adf = pd.read_csv('advertisement.csv')
adf
```

```python
adf['married'] = adf['married'].astype(int)
adf['gender'] = adf['gender'].map({'Male': 1, 'Female': 0})
adf
```

```python
from sklearn.preprocessing import OneHotEncoder 

# Create an instance of One-hot-encoder 
enc = OneHotEncoder() 
  
# Passing encoded columns 
  
enc_data = pd.DataFrame(enc.fit_transform( 
    adf[['occupation', 'education', 'most bought item']]).toarray()) 

adf
```

```python
# Merge with main 
newdf = adf.join(enc_data) 
newdf.drop('city', axis=1, inplace=True)
# adf.drop('education',  axis=1, inplace=True)
# adf.drop('occupation', axis=1, inplace=True)
# adf.drop('most bought item', axis=1, inplace=True)

newdf
```

```python
from sklearn.preprocessing import LabelEncoder
labels_list = []
for lab in adf.labels:
    labels_list.append(lab)

label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels_list)

# print("Encoded Labels:", encoded_labels)
# decoded_labels = label_encoder.inverse_transform(encoded_labels)
# print("Decoded Labels:", decoded_labels)

# adf.drop('education',  axis=1, inplace=True)
adf["encoded_label"] = encoded_labels

# adf.columns
```

```python
# newdf.drop('education',  axis=1, inplace=True)
# newdf.drop('city',  axis=1, inplace=True)
# newdf.drop('most bought item',  axis=1, inplace=True)
# newdf.drop('labels',  axis=1, inplace=True)
# newdf.drop('occupation',  axis=1, inplace=True)

newdf
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split
```

<H3> POWERSET </H3>

```python
# dividing data into target and feature variables

X = newdf
Y = encoded_labels

```

```python
# splitting data into test and training set
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state= 1)
```

```python
# building the decision tree

# making the classifier object
dclf = DecisionTreeClassifier()

# training the classifier
dclf = dclf.fit(xtrain.values, ytrain)

#  predicting labels for test set
ypred = dclf.predict(xtest.values)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# print(dclf.score(xtrain.values, ytrain))
print("Accuracy:", accuracy_score(ytest, ypred))
```

```python
criterion = ["gini", "entropy"]
maxdepth = [3,5,10,20,30]
maxfeatures = [3,5,7,9,11]
combinations = []
for cri in criterion:
    for dep in maxdepth:
        for feat in maxfeatures:
            dclf = DecisionTreeClassifier(criterion=cri, max_depth=dep, max_features=feat)
            dclf = dclf.fit(xtrain.values, ytrain)
            ypred = dclf.predict(xtest.values)
            acc = accuracy_score(ytest, ypred)
            f1mac = f1_score(ytest, ypred, average = "macro")
            f1mic = f1_score(ytest, ypred, average = "micro")

            combinations.append([cri, dep, feat, acc, f1mac, f1mic])
            print("criterion: ", cri, end=' ')
            print("maxdepth: ", dep, end=' ')
            print("maxfeatures: ", feat, end=' ')
            print("f1micro: ", f1mic, end=' ')
            print("f1macro: ", f1mac, end=' ')
            print("accuracy: ", acc)

print(combinations)
```

```python

combinations.sort(key=lambda x: x[4], reverse=True)
for i in range(min(3, len(combinations))):
    print(f"Combination {i + 1}: criterion={combinations[i][0]}, maxdepth={combinations[i][1]}, maxfeatures={combinations[i][2]}, f1micro={combinations[i][3]}, f1macro={combinations[i][4]}, accuracy={combinations[i][5]}")
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, ypred)
```
